```python


#Гистограмма для числовой переменной average_montly_hours:


sns.histplot(df['average_montly_hours'], kde=True)
plt.title('Распределение average_montly_hours')
plt.xlabel('average_montly_hours')
plt.ylabel('Частота')
plt.show()
     
```

```python


#Столбчатая диаграмма для категориальной переменной left:


sns.countplot(x='left', data=df)
plt.title('Распределение left')
plt.xlabel('left')
plt.ylabel('Количество')
plt.show()
     
```

```python


#Боксплот для числовой переменной satisfaction_level:

sns.boxplot(x=df['satisfaction_level'])
plt.title('Boxplot satisfaction_level')
plt.xlabel('satisfaction_level')
plt.show()
     
```



⏰

Распределение average_monthly_hours:
    Данное распределение явно не нормальное. Мы видим два пика — около 150 и 250 часов. Для нормального распределения должен быть один пик в центре и симметричный спад по обе стороны. В данном случае распределение несимметрично и содержит выбросы.

Распределение left:
    Это распределение также не является нормальным, так как оно категориальное и отображает количество людей, которые либо остались в компании (0), либо ушли (1). Нормальное распределение предполагает непрерывные данные, а здесь мы имеем дело с двумя категориями.

Boxplot satisfaction_level:
    По этому графику можно сказать, что данные распределены достаточно равномерно. Боксплот показывает, что выбросы отсутствуют, а само распределение довольно симметричное, хотя не обязательно нормальное.


```python
df: pd.DataFrame = df.drop('department', axis=1)
df: pd.DataFrame = df.drop('salary', axis=1)
# Корреляционная матрица
correlation_matrix: pd.DataFrame = df.corr()

# Визуализация тепловой карты
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Корреляционная матрица')
plt.show()
```

Сильная положительная корреляция:
    number_project и average_montly_hours
    last_evaluation и number_project

Сильная отрицательная корреляция:
    satisfaction_level и left

Практически некоррелированные переменные:
    Work_accident
    promotion_last_5years

Потенциально исключить из анализа можно:
    Work_accident и promotion_last_5years
